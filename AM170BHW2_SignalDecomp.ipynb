```python

data = df.to_numpy()


ica = FastICA(n_components=4)
Sica = ica.fit_transform(data) 
plt.figure(figsize = (20,4))
plt.plot(xSpace, Sica[:,0],label = "source1")
plt.plot(xSpace, Sica[:,1],label = "source2")
plt.plot(xSpace, Sica[:,2],label = "source3")
plt.plot(xSpace, Sica[:,3],label = "source4")
# plt.plot(xSpace, Sica[:,3],label = "source5")
plt.legend()
plt.title('ICA Transform')

```
